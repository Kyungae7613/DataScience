   ### 2 training / validation set 생성

```python
# create training and validation splits

df_train1 = red_wine.sample(frac=0.7, random_state=0)
```

```python
df_valid1 = red_wine.drop(df_train1.index)
```

### 3. training / validation data scaling

```python
# scale to [0, 1]

max_ = df_train1.max(axis=0)
min_ = df_train1.min(axis=0)

df_train2 = (df_train1 - min_) / (max_ - min_)
df_valid2= (df_valid1 - min_) / (max_ - min_)

# split features and target
X_train = df_train2.drop('quality', axis=1)
X_valid = df_valid2.drop('quality', axis=1)
y_train = df_train2['quality']
y_valid = df_valid2['quality']
```

```python
import matplotlib as mpl
import matplotlib.pylab as plt

plt.subplot(1,2,1)
plt.hist(df_train1, bins=1)
plt.subplot(1,2,2)
plt.hist(df_train2, bins=1)

plt.show()
```

### Model & node 생성

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(512, activation='relu', input_shape=[11]),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1)
])
```

###  Model 컴파일

```python
model.compile(
    optimizer='adam',
    loss='mae'
)
```

###  Model 학습

```python
history = model.fit(
            X_train, y_train, 
            validation_data=(X_valid, y_valid), batch_size=256, epochs=10)
```
